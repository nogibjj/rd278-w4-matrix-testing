# Importing Polars

In this assigment Polars are used to import the file, two functions are called from the module source_code to have descriptives and a plot.

```python
import polars as pl
import lib
```

### Importing Data

Inside source_code I created two functions:

1. **descriptive_statistics()**:\
    Parameters:\
        data: Polar DataFrame\
        returns a Polar DataFrame with descriptive statistics\
    
    
2. **generating_plot()**:\
    Parameters:\
        data: Polar DataFrame\
        x_variable: String of the variable in axis X from data\
        y_variable: String of the variable in axis Y from data\
        size: String if a variable is required to add size\
        title: String for the plot title\
        returns a .png file\
    

## Descriptive Statistics

```python
data,results=lib.descriptive_statistics("data/median-income-by-country-2023.csv")
results
```
